```python
#Attach a lakehouse and exeute this code to load data for this question.

import pyspark.sql.functions as F
from notebookutils import mssparkutils

# Use the absolute Fabric pathing syntax for the default attached Lakehouse
csv_path = "Files/Sales"

# Create a dummy dataframe with 1 million rows
df_large = spark.range(0, 1000000) \
    .withColumn("Name", F.concat(F.lit("User_"), F.col("id"))) \
    .withColumn("Age", (F.rand() * 60 + 18).cast("int")) \
    .withColumn("Salary", (F.rand() * 100000).cast("double")) \
    .withColumn("JoinDate", F.date_add(F.current_date(), -1 * (F.rand() * 365).cast("int")))

# Ensure target folder is clear via mssparkutils to prevent metadata conflicts
try:
    mssparkutils.fs.rm(csv_path, True)
except:
    pass

# Write out the CSV files directly
df_large.write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(csv_path)

print("Data written successfully!")
```

```python
# Clear cache to ensure a fair test
spark.catalog.clearCache()

# Measure tine taken for reading without inferschema
start_time = time.time()
df_no_infer = spark.read.option("header", "true").csv("Files/Sales")
end_time = time.time()
print(f"Time taken WITHOUT inferSchema: {end_time - start_time:.4f} seconds")

# Display the schema to check the data types
df_no_infer.printSchema()

# Clear cache to ensure a completely clean run
spark.catalog.clearCache()

# Measure tine taken for reading with inferschema
start_time = time.time()
df_infer = spark.read.option("header", "true").option("inferSchema", "true").csv("Files/Sales")
end_time = time.time()
print(f"Time taken WITH inferSchema: {end_time - start_time:.4f} seconds")

# Display the schema to see the discovered data types
df_infer.printSchema()
```
