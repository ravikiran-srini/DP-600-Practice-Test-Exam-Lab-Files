```python
# Option A

from pyspark.sql.functions import col

data = [
    # 2024: 3 Months
    (1, 22.5, 2024, 1, 1), (2, 21.0, 2024, 1, 2), # Jan: 2 days
    (3, 19.5, 2024, 2, 14),                      # Feb: 1 day
    (4, 25.2, 2024, 3, 10), (5, 24.8, 2024, 3, 11), # Mar: 2 days

    # 2025: 4 Months
    (6, 15.1, 2025, 1, 1),                       # Jan: 1 day
    (7, 14.0, 2025, 2, 5),  (8, 13.5, 2025, 2, 6),  # Feb: 2 days
    (9, 18.9, 2025, 3, 20),                      # Mar: 1 day
    (10, 20.1, 2025, 4, 12), (11, 20.5, 2025, 4, 13) # Apr: 2 days
]

columns = ["SensorID", "Temperature", "year", "month", "day"]
df = spark.createDataFrame(data, columns)

df.write \
    .mode("overwrite") \
    .parquet("Files/WeatherTelemetry")

print("WeatherTelemetry data written successfully.")
```

```python
# Option B

# Read the data back
df_read = spark.read.parquet("Files/WeatherTelemetry")

# Check how many RDD partitions Spark has created for this data
print(f"Number of Spark partitions: {df_read.rdd.getNumPartitions()}")
```

```python
# Option D

from pyspark.sql import SparkSession
import time

spark = SparkSession.builder.getOrCreate()

# Sample data
data = [
    (2026, 5, 11, 35.5),
    (2026, 5, 11, 36.1),
    (2026, 5, 12, 34.8),
    (2026, 6, 1, 38.2)
]

# Create DataFrame
df = spark.createDataFrame(
    data,
    ["year", "month", "day", "temperature"]
)
```

```python
# Option D

# Write WITH partitioning
start = time.time()

df.write \
    .partitionBy("year", "month", "day") \
    .mode("overwrite") \
    .parquet("Files/WeatherTelemetryPartitioned")

end = time.time()

print("Partitioned write time:", end - start)
```

```python
# Option D

# Write WITHOUT partitioning
start = time.time()

df.write \
    .mode("overwrite") \
    .parquet("Files/WeatherTelemetryNonPartitioned")

end = time.time()

print("Non-partitioned write time:", end - start)
```
