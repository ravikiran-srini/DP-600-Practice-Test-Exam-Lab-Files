```python
from pyspark.sql.types import StructType, StructField, StringType, TimestampType
from pyspark.sql.functions import col
from datetime import datetime

# 1. Define the schema with the proper Timestamp data type
schema = StructType([
    StructField("TrackingNumber", StringType(), True),
    StructField("PackageWeight", StringType(), True),       
    StructField("ShipmentTimestamp", TimestampType(), True), # Set as proper Timestamp
    StructField("DeliveryStatus", StringType(), True)
])

# 2. Provide data using proper Python datetime objects
data = [
    ("TRK1001", "12.50", datetime(2026, 6, 1, 10, 30, 0), "Delivered"),
    ("TRK1002", "4.25",  datetime(2026, 6, 1, 14, 15, 0), "In Transit"),
    ("TRK1003", "45.00", datetime(2026, 6, 2, 9, 0, 0),   "Pending")
]

# 3. Load the data into the initial DataFrame
df = spark.createDataFrame(data, schema=schema)

# 4. Display the final result
display(df)
```

```python
# 5. Transform data using exactly TWO operations
df = df.select("PackageWeight", col("PackageWeight").cast("double")) \
.withColumn("ShipmentDate", df["ShipmentTimestamp"].cast("date"))

# 6. Display the final result
display(df)
```
