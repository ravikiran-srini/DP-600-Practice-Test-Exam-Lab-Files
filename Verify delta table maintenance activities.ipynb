```python
# Create a simple DataFrame and write it to a Delta table named 'SalesPerson'
data1 = [("S001", "Alice", "New York"), ("S002", "Bob", "London")]
df1 = spark.createDataFrame(data1, ["SalesPersonID", "Name", "City"])
df1.write.format("delta").mode("overwrite").saveAsTable("SalesPerson")

# Append a second batch of data (creates an additional small file)
data2 = [("S003", "Charlie", "Paris")]
df2 = spark.createDataFrame(data2, ["SalesPersonID", "Name", "City"])
df2.write.format("delta").mode("append").saveAsTable("SalesPerson")

# Append a third batch of data (creates another small file)
data3 = [("S004", "David", "Tokyo")]
df3 = spark.createDataFrame(data3, ["SalesPersonID", "Name", "City"])
df3.write.format("delta").mode("append").saveAsTable("SalesPerson")
```

```python
%%sql
EXPLAIN TABLE SalesPerson
```

```python
%%sql
REFRESH TABLE SalesPerson
```

```python
%%sql
DESCRIBE HISTORY SalesPerson;
```

```python
%%sql
DESCRIBE DETAIL SalesPerson;
```

```python
%%sql
-- 1. Compact the small files
OPTIMIZE SalesPerson;
```

```python
%%sql
DESCRIBE HISTORY SalesPerson;
```

```python
%%sql
DESCRIBE DETAIL SalesPerson;
```

```python
%%sql
-- 2. Disable the retention duration check so we can vacuum immediately
SET spark.databricks.delta.vacuum.parallelDelete.enabled = true;
SET spark.databricks.delta.retentionDurationCheck.enabled = false;

-- 3. Clean up the old, unoptimized files (retaining 0 hours of history)
VACUUM SalesPerson RETAIN 0 HOURS;
```
